```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Cargando data del 1 al 5 de Abril
abril_01_df = pd.read_csv('validacionTroncal20230401.csv') #Cargando data Primero de Abril 
abril_02_df = pd.read_csv('validacionTroncal20230402.csv')
abril_03_df = pd.read_csv('validacionTroncal20230403.csv')
abril_04_df = pd.read_csv('validacionTroncal20230404.csv')
abril_05_df = pd.read_csv('validacionTroncal20230405.csv')
```

```python
#Concatenando data 
transmi_df=pd.concat([abril_01_df,abril_02_df,abril_03_df, abril_04_df, abril_05_df])
transmi_df
```

```python
transmi_df.dtypes
```

```python
#SELECCIÓN DE COLUMNAS CORRELACIONADAS
transmi_df = transmi_df[['Acceso_Estacion',
 'Day_Group_Type',
 'Estacion_Parada',
 'Fecha_Clearing',
 'Fecha_Transaccion',
 'Linea',
 'Nombre_Perfil',
 'Tipo_Tarjeta',
 'Valor'
]]
#method=bfill
#aeropuertos_df=aeropuertos_df.fillna() 
transmi_df.tail()
```

```python
#CAMBIANDO NaN POR 0 
transmi_df=transmi_df.fillna(0) 
transmi_df.tail()
```

```python
#Agregando fila con día del mes 

transmi_df['Dia']=transmi_df['Fecha_Clearing'].str.slice(9, 10)
transmi_df
```

```python
#Agregando fila con hora de transacción 

transmi_df['Hora']=transmi_df['Fecha_Transaccion'].str.slice(11, 13)
transmi_df
```

```python
#Descargando Data para trabajar desde PowerBi la visualización
transmi_df.to_csv('Data_transmilenio.csv', index=False)
```

## RESUMEN MESES DESDE OCTUBRE A DICIEMBRE 2022 Y ENERO A ABRIL 2023

```python
resumen_oct22 = pd.read_excel('Resumen_oct.xlsx')
resumen_nov22 = pd.read_excel('Resumen_nov.xlsx')
resumen_dic22 = pd.read_excel('Resumen_dic.xlsx')

resumen_ene23 = pd.read_excel('Resumen_enero.xlsx')
resumen_feb23 = pd.read_excel('Resumen_feb.xlsx')
resumen_mar23 = pd.read_excel('Resumen_mar.xlsx')
resumen_abr23 = pd.read_excel('Resumen_abril.xlsx')



#Concatenando data 
resumen_meses=pd.concat([resumen_oct22,resumen_nov22,resumen_dic22, resumen_ene23, resumen_feb23,resumen_mar23,resumen_abr23])
resumen_meses
```

```python
#CAMBIANDO NaN POR 0 
resumen_meses=resumen_meses.fillna(0) 
resumen_meses.tail()
```

```python
resumen_meses = resumen_meses.drop(resumen_meses[resumen_meses['Acceso de Estación']=='(0) (Unknown)'].index) #Quitando datos de acceso a estación desconocidos
```

```python
resumen_meses = resumen_meses.drop(resumen_meses[resumen_meses['Línea']=='(1)'].index) #Quitando datos de acceso a estación desconocidos
```

```python
resumen_meses = resumen_meses.drop(resumen_meses[resumen_meses['Acceso de Estación']=='Total general'].index) #Quitando datos de acceso a estación desconocidos
```

```python
resumen_meses = resumen_meses.drop(resumen_meses[resumen_meses['Acceso de Estación']==0].index) #Quitando datos de acceso a estación desconocidos
```
